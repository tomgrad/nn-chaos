# Odwzorowanie logistyczne (mapa logistyczna)

Jednowymiarowe odwzorowanie unimodalne z czasem dyskretnym.
- określone na skończonym odcinku
- posiada pojedyńcze gładkie maksimum
- nieodwracalne

$$
x_{n+1} = r x_n (1-x_n)
$$

gdzie $r$ jest parametrem $r \in [0,4]$

```julia
using PyPlot
logmap(x, r) = r*x*(1-x)
x=0:0.02:1
plot(x,logmap.(x,4))
# plot([[0,0],[1,1]])
xlabel(L"x_n")
ylabel(L"x_{n+1}");
```

```julia
function run(r)
    x=0.1
    Nt=1000
    for i in 1:Nt
        x=logmap.(x,r)
    end

    X=[]
    N=100
    for i in 1:N
        x=logmap.(x,r)
        append!(X,x)
    end
    X
end
```

```julia
# r: 2.8 (1), 3.4 (2), 3.5 (4), 3.56 (8), 3.565 (16), 3.6 (c), 4 (c)
x=run(2.8)
figure(figsize=(10,4))
subplot(121)
plot(x[1:20],".-")
title(L"x(t)")
subplot(122)
title(L"x_{n+1}(x_n)")
plot(x[1:end-1],x[2:end],"+");
```

```julia
figure(figsize=(9, 6))
xlabel("r")
ylabel("x")

for r in 2.8:0.001:4
    x=run(r)
    plot([r for i in x],x,"k,",alpha=0.3)
end
```
